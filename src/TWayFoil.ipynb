```julia
# %ignore-cell # !„Äê2024-01-22 00:10:15„ÄëÂÆûÈôÖ‰ΩøÁî®ÁöÑÂ∫ì‰ª£Á†ÅÂπ∂‰∏çÈúÄË¶ÅËøôÊ†∑ÁöÑ„ÄåHEX„ÄçÁºñÁ†Å
HEX(argb::ARGB)::UInt32 = (
    + (UInt32(round(argb.alpha * 0xff)) << 24)
    + (UInt32(round(argb.r     * 0xff)) << 16)
    + (UInt32(round(argb.g     * 0xff)) << 8)
    + (UInt32(round(argb.b     * 0xff)))
)

HEX(color)::UInt32 = HEX(ARGB(color))

HEX2ARGB(hex::UInt32)::ARGB = ARGB(
    ((hex >> 16) & 0xff) / 0xff,
    ((hex >> 8 ) & 0xff) / 0xff,
    ((hex      ) & 0xff) / 0xff,
    ((hex >> 24) & 0xff) / 0xff,
)

let p = test[0x80, 0x40]
    @show p HEX(p)
    p2 = HSV(0.5, 0.5, 0.5)
    p3 = HEX2ARGB(0x8066ccff)
    @show HEX(p2) HEX(p3) p3
end
```

```julia
"""
ËØÜÂà´Âπ∂Ë£ÅÂâ™PNG„ÄåËá™Âä®Ë°•Ë∂≥„ÄçÁöÑÂ≠óËäÇ
- üìåÊåâÁÖß„ÄêÂõ∫ÂÆöÊ∑ªÂä†„ÄëÁöÑÊú´Â∞æÂ≠óËäÇÔºåËÆ°ÁÆó„ÄåÂÖàÂâçËΩ¨Êç¢ÊàêARGB PNGÊó∂ÔºåÂõ†„Äé4Â≠óËäÇÈúÄË¶Å„ÄèËÄå„ÄêÂ§öÊñ∞Â¢û„ÄëÁöÑÂ≠óËäÇÊï∞„Äç
    - Ê≠§Â§Ñ‰ªÖÈúÄ‰∏Ä‰∏™Â≠óËäÇÔºåÂπ∂‰∏îËøô‰∏™Â≠óËäÇÂõ∫ÂÆöÂú®ÊúÄÂêé
    - ÊâÄÊúâÂèØËÉΩÂèñÂÄº‰∏∫Ôºö0x01~0x04
        - ‰∏ãÈôê/0x01ÔºöÊ∑ªÂä†ËØ•Â≠óËäÇÂêéÊ≠£Â•ΩÔºåÂè™ÈúÄÂà†ÊéâÊú¨Ë∫´ÔºàÂ∑≤ÁªèÁÆóÂÖ•Âú®pop!ÈáåËæπ‰∫ÜÔºâ
        - ‰∏äÈôê/0x04ÔºöÊ∑ªÂä†ËØ•Â≠óËäÇÂêéÂÜó‰Ωô3‰∏™Â≠óËäÇÔºåËØ•Â≠óËäÇÂíåÂÖ∂ÂÆÉ‰ª®Ôºà‰∏ÄËà¨‰∏∫Á©∫ÔºâÁªÑÊàê‰∫ÜÊñ∞ARGBÂÉèÁ¥†
    - ÂÖ∂ÂÆÉËåÉÂõ¥Â§ñÂèñÂÄºÔºöËøîÂõûËá™Ë∫´
"""
function strip_raw_by_last_pixel!(raw8::Vector{UInt8})
    # * Âà©Áî®„ÄêÂõ∫ÂÆöÊ∑ªÂä†„ÄëËé∑ÂèñÊú´Â∞æ„ÄåÊñ∞Â¢ûÊï∞„Äç
    local num_surplussed_bytes = raw8[end]
    # ‰ΩøÁî®„ÄåÊñ∞Â¢ûÊï∞+1„ÄçËøõË°å„ÄêÊà™Âèñ„Äë
    if num_surplussed_bytes in 0x01:0x04
        for _ in 1:num_surplussed_bytes
            pop!(raw8)
        end
    end
    # ËøîÂõû
    raw8
end

"ÈÄÇÂ∫îÂÖ∂ÂÆÉÁ±ªÂûãÔºåÂ¶ÇÁîüÊàêÂô®"
strip_raw_by_last_pixel(raw8) = raw8 |> collect |> strip_raw_by_last_pixel!

# %ignore-begin # * ÊµãËØï
let 
    s1 = @show strip_raw_by_last_pixel(UInt8(c) for c in "12345") # ! „ÄåUInt('5') == 0x35„ÄçËøáÂ§ßÔºå‰∏çÂÅöÂ§ÑÁêÜ
    s2 = @show strip_raw_by_last_pixel(UInt8(c) for c in "12345\x01") # ÂêéÁª≠Ëøô‰ø©Ê≠£Â∏∏Êìç‰Ωú
    s3 = @show strip_raw_by_last_pixel(UInt8(c) for c in "12345\x00\x00\x00\x04") # ÂêéÁª≠Ëøô‰ø©Ê≠£Â∏∏Êìç‰Ωú
    @assert s1 == s2 == s3
end
# %ignore-end

#= %only-compiled
export png2raw # * ÂØºÂá∫Ê†∏ÂøÉÂáΩÊï∞
%only-compiled =#

"""
    png2raw(ÁõÆÊ†áÁ±ªÂûãorÁõÆÊ†áË∑ØÂæÑ, ÂæÖËΩ¨Êç¢ÂØπË±°)
PNG‚áíÊñá‰ª∂
- üìåÂ∞ÜPNGÊåâÁÖß„ÄåÂÖàË°åÂêéÂàó„ÄçÁöÑÊñπÂºèÊâ´ÊèèÔºåÂπ∂ËøîÂõûÂ≠óËäÇ‰∏≤
- üìåÊ†∏ÂøÉÊµÅÁ®ãÔºöÂõæÁâáË∑ØÂæÑ‚áíÂÉèÁ¥†Áü©Èòµ‚áíÂ≠óËäÇ‰∏≤‚áíÊñá‰ª∂
"""
function png2raw end

# ÂõæÁâá‚áíÁü©Èòµ
png2raw(from::AbstractString) = png2raw(
    replace(from, r".png$" => ""),
    from
)
png2raw(destination, from::AbstractString) = png2raw(
    destination,
    PNGFiles.load(from) # Âä†ËΩΩÂá∫‰∏Ä‰∏™Áü©Èòµ
)

# Áü©Èòµ‚áíÂêëÈáè
function png2raw(png::Matrix)
    # * ÂºÄÂßãËÆ°ÁÆó„ÄÅÂ≠òÂÇ®Âπ∂ËøîÂõûARGBÂ∫èÂàó * #
    local raw8 = UInt8[]
    local argb::ARGB

    # * ÂÖàÊ≥®ÂÖ•ÈïøÂ∫¶
    for row in eachcol(png)
        for pixel in row
            # Âä®ÊÄÅËé∑ÂèñÂÉèÁ¥†
            argb = ARGB(pixel)
            # ‰∏Ä‰∏™ÂÉèÁ¥†ËΩ¨Êç¢ÊàêÂõõ‰∏™Â≠óËäÇ
            push!(raw8,
                UInt8(round(argb.alpha * 0xff)),
                UInt8(round(argb.r     * 0xff)),
                UInt8(round(argb.g     * 0xff)),
                UInt8(round(argb.b     * 0xff)),
            )
        end
    end
    
    # * ÂêéËøõË°åÊà™Âèñ
    return strip_raw_by_last_pixel!(raw8)
end
png2raw(destination, png::Matrix) = png2raw(destination, #= Áü©Èòµ‚áíÂêëÈáè =#png2raw(png))

# ÂêëÈáè‚áíÊñá‰ª∂|ÂêëÈáè
png2raw(::Type{<:Vector}, raw::Vector) = raw # * ÂêëÈáèÁ±ªÂûã‚áíËá™Ë∫´
png2raw(destination::AbstractString, raw::Vector) = write(
    destination,
    raw
)

# %ignore-begin
let raw = png2raw(test)
    # Â∞ùËØïÊääËØªÂèñÂêéÁöÑÊµãËØïÂõæÁâáËøõË°åËΩ¨Êç¢
    @time @show write("png2raw(test).test.txt", raw)
    raw
end
# %ignore-end
```

Â≠óËäÇ‰∏≤‚áíÂÉèÁ¥†Â∫èÂàó

```julia
"""
ËÆ°ÁÆó„ÄåËá™Âä®Ë°•Ë∂≥„ÄçËøáÁ®ã‰∏≠ÈúÄË¶ÅË°•Ë∂≥ÁöÑ„ÄåÁ©∫ÁôΩÂ≠óËäÇ„Äç‰∏™Êï∞
- üìåÈô§‰∫ÜÊúÄÂêé‰∏Ä‰∏™Ë°®Á§∫„ÄåËΩ¨Êç¢Êó∂Ë°•Ë∂≥ÁöÑÂ≠óËäÇ‰∏™Êï∞„ÄçÁöÑÂ≠óËäÇÔºåÂÖ∂ÂÆÉÈÉΩ‰∏∫„ÄåÁ©∫ÁôΩÂ≠óËäÇ„Äç
    - ‚ö†Ô∏èÂÖ∂Êï∞ÁõÆÂèØËÉΩ‰∏∫Èõ∂
"""
n_complement_blank_bytes(len::Integer) = 3 - (len & 0b11)

"""
Êõ¥ÈÄöÁî®Âú∞ËÆ°ÁÆó„ÄåËá™Âä®Ë°•Ë∂≥nÂ≠óËäÇÂçïÂÖÉ„ÄçÊâÄÈúÄÁöÑ„ÄåÁ©∫ÁôΩÂ≠óËäÇÊï∞„Äç
"""
n_complement_blank_bytes(len::Integer, n::Unsigned) = (n - 1) - (len & 0b11)

# %ignore-begin
let
    @assert n_complement_blank_bytes.([
        12
        13
        14
        15
    ]) == [
        3
        2
        1
        0
    ]
end
# %ignore-end

"""
Ëá™Âä®Ë°•Ë∂≥
- üìåÊåâÁÖß`png2raw`ÁöÑÊÄùË∑ØÔºåÂØπÂ≠óËäÇ‰∏≤ËøõË°å„ÄåÂõõ‰ΩçË°•Ë∂≥„Äç
    - ‰∏ÄÂÆö‰ºö„ÄêË°•Ë∂≥„Äë‰∏Ä‰∏™ÂÉèÁ¥†ÔºåÊïÖÂ≠óËäÇÂÄºÂøÖÂÆöÂ§ß‰∫éÈõ∂
"""
function complete_4_bytes!(raw::AbstractVector{UInt8})
    # Ê†πÊçÆ‰∏™Êï∞Ë°•Ë∂≥„ÄåÁ©∫ÁôΩÂ≠óÁ¨¶„Äç
    local n_blank_bytes::UInt8 = n_complement_blank_bytes(length(raw))
    for _ in 1:n_blank_bytes
        push!(raw, UInt8(0))
    end

    # Ê∑ªÂä† | ËøôÊó∂ÂÄôË¶ÅÁÆó‰∏äÂÖ∂Ëá™Ë∫´
    push!(raw, UInt8(n_blank_bytes + 1))

    # ËøîÂõûÁ©∫ | Êñ≠Ë®Ä`@assert (length(v) & 0b11) == 0b0 raw`‰∏ÄÂÆöÊàêÁ´ã
    nothing
end

# %ignore-begin
let v = UInt8[1,2,3,4,5]
    complete_4_bytes!(v)
    @assert (length(v) & 0b11) == 0b0 raw
    complete_4_bytes!(v)
    @assert (length(v) & 0b11) == 0b0 raw
    push!(v, 1, 2)
    complete_4_bytes!(v)
    @assert (length(v) & 0b11) == 0b0 raw
    push!(v, 1, 2, 3)
    complete_4_bytes!(v)
    @assert (length(v) & 0b11) == 0b0 raw
    @show v
end
# %ignore-end

"""
Â≠óËäÇ‰∏≤‚áíÂÉèÁ¥†Â∫èÂàó
- üìåÈ¶ñÂÖàËøõË°å„ÄåËá™Âä®Ë°•Ë∂≥„ÄçÔºàÁ†¥ÂùèÊÄßÊìç‰ΩúÔºÅÔºâ
- ÁÑ∂Âêé„ÄåÂõõ‰ΩçËΩ¨Êç¢„Äç
"""
function raw2argb!(raw::AbstractVector{UInt8})::Vector{ARGB{N0f8}}
    # Ëá™Âä®Ë°•Ë∂≥
    complete_4_bytes!(raw)
    # Âõõ‰∏™‰∏ÄÊâπËΩ¨Êç¢
    local result = ARGB{N0f8}[] # !„Äê2024-01-21 23:54:57„ÄëÂêéÁª≠ÂøÖÈ°ªÂæóÊòØ`Matrix{ARGB{N0f8}}`Á±ªÂûãÊâçËÉΩË¢´Â≠òÂÇ®
    for i in 0:((length(raw) >> 2 #= ÂÉèÁ¥†‰∏™Êï∞ =#) - 1)
        push!(result, ARGB{N0f8}(
            reinterpret(N0f8, raw[(i << 2) + 2]), # R
            reinterpret(N0f8, raw[(i << 2) + 3]), # G
            reinterpret(N0f8, raw[(i << 2) + 4]), # B
            reinterpret(N0f8, raw[(i << 2) + 1]), # A # ! Âú®ÊúÄÂâçÈù¢
        ))
    end
    result
end

# %ignore-begin
let
    @show raw2argb!(copy(file_raw))
end
# %ignore-end
```

ÂÉèÁ¥†Â∫èÂàó‚áíÂÉèÁ¥†Áü©Èòµ

```julia
"""
ÂàÜËß£Êï¥Êï∞‰∏∫„Äå‰∏§‰∏™Áõ∏ËøëÊï¥Êï∞ÁöÑ‰πòÁßØ„Äç
- üéØÊ†πÊçÆÂÉèÁ¥†Â∫èÂàóËÆ°ÁÆóÊúÄÁªàÂõæÂÉèÂ∞∫ÂØ∏
    - ÂéüÂàôÔºöÂõæÂÉèÂ∞ΩÂèØËÉΩ„ÄêÊñπÊ≠£„ÄëËÄåÈùû„ÄêÊâÅÈïø„Äë
- üìå‰ΩøÁî®Ë¥®Âõ†Êï∞ÂàÜËß£Ôºå‰ªéÂºÄÊ†πÂæÄÂ∞èÂ§ÑËøáÊª§
    - Á¨¨‰∏Ä‰∏™Êï∞ ‚â§ Á¨¨‰∫å‰∏™Êï∞
"""
function size_1to2(n::I)::Tuple{I,I} where {I <: Integer}
    # È¶ñÂÖàÂºÄÊ†π
    local n1::I = I(floor(sqrt(n)))
    # ÁÑ∂ÂêéÈÄíÂáè
    while !iszero(n % n1) && n1 > 0
        n1 -= 1
    end
    # ÊúÄÂêéÁõ∏Èô§ËøîÂõû
    return (n1, n √∑ n1)
end

# %ignore-begin
let
    @assert @show(size_1to2(12)) == (3, 4)
    @assert @show(size_1to2(123)) == (3, 41)
    @assert @show(size_1to2(169)) == (13, 13)
end
# %ignore-end

"""
ÈáçÊï¥Âåñ„ÄåÂÉèÁ¥†ÂêëÈáè„Äç
- üìåÈúÄË¶Å‰øùÁïôÂÖ∂‰∏≠ÂÉèÁ¥†ÁöÑÁ±ªÂûã
    - Âè™ÊúâÁ±ª‰ºº`Matrix{ARGB{N0f8}}`ÁöÑÁ±ªÂûãÊâçËÉΩË¢´save
"""
reshape_pixels(pixels::Vector{T}) where T = Matrix{T}(reshape(pixels, size_1to2(length(pixels))))

# %ignore-begin
let
    @show file_raw |> copy |> raw2argb! |> reshape_pixels
end
# %ignore-end
```

```julia
#= %only-compiled
export raw2png!, raw2png # * ÂØºÂá∫Ê†∏ÂøÉÂáΩÊï∞
%only-compiled =#

"""
    raw2png(ÁõÆÊ†áÁ±ªÂûãorÁõÆÊ†áË∑ØÂæÑ, ÂæÖËΩ¨Êç¢ÂØπË±°)
Â∞ÜÊñá‰ª∂ËΩ¨Êç¢‰∏∫ÂõæÁâá
- üìå‰∏ªË¶ÅÊµÅÁ®ãÔºöÊñá‰ª∂Ë∑ØÂæÑ‚áíÂ≠óËäÇ‰∏≤ÔºàÂêëÈáèÔºâ‚áíÂÉèÁ¥†Áü©ÈòµÔºàÂõæÂÉèÔºâ|ÂõæÁâáË∑ØÂæÑ
"""
function raw2png end

# Êñá‰ª∂‚áíÂêëÈáè
raw2png(from::AbstractString) = raw2png("$from.png", from)
raw2png(destination, from::AbstractString) = raw2png(
    destination,
    from |> read
)

# ÂêëÈáè‚áíÁü©Èòµ
raw2png!(::Type{<:Vector}, raw::Vector{UInt8}) = raw # * Ëã•ÊåáÂÆöËæìÂá∫Ê†ºÂºè‰∏∫„ÄåÂêëÈáè„ÄçÔºåÂàôËøîÂõûËá™Ë∫´
raw2png!(destination, raw::Vector{UInt8}) = raw2png(
    destination,
    raw |> raw2argb! |> reshape_pixels
)
raw2png(destination, raw::Vector{UInt8}) = raw2png!(destination, copy(raw))

# Áü©Èòµ‚áíËá™Ë∫´|Êñá‰ª∂
raw2png(::Type{<:Matrix}, image::Matrix) = image # * Ëã•ÊåáÂÆöËæìÂá∫Ê†ºÂºè‰∏∫„ÄåÁü©Èòµ„ÄçÔºåÂàôËøîÂõûËá™Ë∫´
raw2png(destination::AbstractString, image::Matrix) = PNGFiles.save(destination, image)

# %ignore-below # * ÊµãËØï
let # * Á∫ØÊµãËØïÂõæÁâá
    raw2png("$file.test.png", file)
    raw2png(path) # ! Â∞ùËØïÊääÊµãËØïÁî®ÂõæÁâáÂÜçËΩ¨Êç¢‰∏ã
end
# Â∞ùËØïËá™ÂéãÁº©Ëá™Ëß£ÂéãÔºå‰ª•‰æøËá™Ê£Ä
let SELF_PATH = "TWayFoil.ipynb"
    content = read(SELF_PATH, String)
    png = raw2png(Matrix, SELF_PATH)
    @show size(png)
    raw = png2raw(Vector, png)
    re_content = String(raw)
    if re_content != content
        write("err.re_content.test.txt", re_content)
        write("err.content.test.txt", content)
    end
    @assert re_content == content
end
```
